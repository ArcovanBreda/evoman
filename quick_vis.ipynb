```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load and process data
data = []
with open('./optimization_test_arco/results.txt', 'r') as f:
    for line in f:
        data.append(line.strip().split()[2:6])

generations = list(range(1, len(data) + 1))
best = [float(row[0]) for row in data]
average = [float(row[1]) for row in data]
std = [float(row[2]) for row in data]

# Set plot style
plt.figure(figsize=(12, 7))
sns.set(style='whitegrid')

# Plot 'average' with line and 'std' as a filled area
plt.plot(generations, average, label='Average', color='orange', linestyle='--', marker='s', markersize=6, linewidth=2)
plt.fill_between(generations, [a - s for a, s in zip(average, std)], [a + s for a, s in zip(average, std)], 
                 color='orange', alpha=0.3, label='Standard Deviation')

# Plot 'best' as a separate line
plt.plot(generations, best, label='Best', color='royalblue', linestyle='-', marker='o', markersize=6, linewidth=2)

# Customize labels and title
plt.xlabel('Generation', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Performance Over Generations', fontsize=16)

# Customize ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add a legend
plt.legend(fontsize=12, loc='upper right')

# Add gridlines
plt.grid(True, which='both', linestyle='--', linewidth=0.7)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load and process data
data = []
with open('./optimization_test_arco/results.txt', 'r') as f:
    for line in f:
        data.append(line.strip().split()[2:6])

generations = list(range(1, len(data) + 1))
best = [float(row[0]) for row in data]
average = [float(row[1]) for row in data]
std = [float(row[2]) for row in data]

# Set plot style
plt.figure(figsize=(12, 7))
sns.set(style='whitegrid')

# Create a primary axis for 'average' and 'std'
fig, ax1 = plt.subplots(figsize=(12, 7))

# Plot 'average' on the primary axis
ax1.plot(generations, average, label='Average', color='orange', linestyle='--', marker='s', markersize=6, linewidth=2)
ax1.fill_between(generations, [a - s for a, s in zip(average, std)], [a + s for a, s in zip(average, std)], 
                 color='orange', alpha=0.3, label='Standard Deviation')

# Customize the primary axis
ax1.set_xlabel('Generation', fontsize=14)
ax1.set_ylabel('Average / Std Deviation', fontsize=14, color='orange')
ax1.tick_params(axis='y', labelcolor='orange')
ax1.set_xticks(generations)
ax1.grid(True, which='both', linestyle='--', linewidth=0.7)

# Create a secondary axis for 'best'
ax2 = ax1.twinx()
ax2.plot(generations, best, label='Best', color='royalblue', linestyle='-', marker='o', markersize=6, linewidth=2)

# Customize the secondary axis
ax2.set_ylabel('Best', fontsize=14, color='royalblue')
ax2.tick_params(axis='y', labelcolor='royalblue')

# Add legends for both axes
ax1.legend(loc='upper left', fontsize=12)
ax2.legend(loc='upper right', fontsize=12)

# Set the title
plt.title('Performance Over Generations', fontsize=16)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def quick_vis(experiment_name="optimization_test_arco5"):

    # Load and process data
    data = []
    with open(f'./{experiment_name}/results.txt', 'r') as f:
        for line in f:
            data.append(line.strip().split()[2:6])

    generations = list(range(1, len(data) + 1))
    best = [float(row[0]) for row in data]
    average = [float(row[1]) for row in data]
    std = [float(row[2]) for row in data]

    # Set plot style
    plt.figure(figsize=(12, 7))
    sns.set(style='whitegrid')

    # Plot 'average' with line and 'std' as a filled area
    plt.plot(generations, average, label='Average', color='orange', linestyle='--', marker='s', markersize=6, linewidth=2)
    plt.fill_between(generations, [a - s for a, s in zip(average, std)], [a + s for a, s in zip(average, std)], 
                    color='orange', alpha=0.3, label='Standard Deviation')

    # Plot 'best' as a separate line
    plt.plot(generations, best, label='Best', color='royalblue', linestyle='-', marker='o', markersize=6, linewidth=2)

    # Customize labels and title
    plt.xlabel('Generation', fontsize=14)
    plt.ylabel('Value', fontsize=14)
    plt.title('Performance Over Generations', fontsize=16)

    # Customize ticks
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    # Add a legend
    plt.legend(fontsize=12, loc='upper right')

    # Add gridlines
    plt.grid(True, which='both', linestyle='--', linewidth=0.7)

    # Adjust layout
    plt.tight_layout()

    # Show the plot
    plt.show()
```
